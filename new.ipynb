```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

conf_matrices = {
    "CNN": np.array([[447, 2], [3, 18]]),
    "Random Forest": np.array([[446, 3], [3, 18]]),
    "XGBoost": np.array([[447, 2], [5, 16]])
    
}

labels = ["Normal", "Porosity"]
fig, axes = plt.subplots(1, 3, figsize=(12, 5))

for ax, (title, matrix) in zip(axes, conf_matrices.items()):
    # Normalize matrix for percentage display
    percentages = matrix / matrix.sum(axis=1, keepdims=True)
    annotations = np.array([f"{val}\n({perc:.1%})" for val, perc in zip(matrix.flatten(), percentages.flatten())]).reshape(2, 2)

    sns.heatmap(matrix, annot=annotations, fmt='', cmap="Blues", cbar=False,
                xticklabels=labels, yticklabels=labels, ax=ax)

    ax.set_title(title)
    ax.set_xlabel("Predicted")
    ax.set_ylabel("True")

plt.suptitle("Confusion Matrices of Supervised Models", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

conf_matrices = {
    "CNN": np.array([[447, 2], [3, 18]]),
    "Random Forest": np.array([[446, 3], [3, 18]]),
    "XGBoost": np.array([[447, 2], [5, 16]])
}

labels = ["Normal", "Porosity"]
fig, axes = plt.subplots(1, 3, figsize=(12, 5))  # Increased figure size

for ax, (title, matrix) in zip(axes, conf_matrices.items()):
    percentages = matrix / matrix.sum(axis=1, keepdims=True)
    annotations = np.array([f"{val}\n({perc:.1%})" for val, perc in zip(matrix.flatten(), percentages.flatten())]).reshape(2, 2)

    # Plot heatmap
    sns.heatmap(
    matrix,
    annot=annotations,
    fmt='',
    cmap="Blues",
    cbar=False,
    xticklabels=labels,
    yticklabels=labels,
    ax=ax,
    annot_kws={"fontsize": 14}  #  Increase this value to make text bigger
)

    ax.set_title(title, fontsize=16)
    ax.set_xlabel("Predicted", fontsize=16)
    ax.set_ylabel("True", fontsize=16)
    ax.tick_params(labelsize=16)

plt.suptitle("Confusion Matrices of Supervised Models", fontsize=18)
plt.tight_layout()
plt.savefig("conf_matrices_supervised.png", dpi=300)  # Save at high resolution
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Confusion matrices for unsupervised models
conf_matrices = {
    "Autoencoder (95th)": np.array([[1457, 36], [28, 43]]),
    "Autoencoder (99th)": np.array([[1489, 4], [59, 12]]),
    "DBSCAN": np.array([[1461, 32], [34, 37]])
}

labels = ["Normal", "Porosity"]
fig, axes = plt.subplots(1, 3, figsize=(12, 5))  # Wider and taller for clarity

for ax, (title, matrix) in zip(axes, conf_matrices.items()):
    # Calculate percentages
    percentages = matrix / matrix.sum(axis=1, keepdims=True)
    annotations = np.array([
        f"{val}\n({perc:.1%})"
        for val, perc in zip(matrix.flatten(), percentages.flatten())
    ]).reshape(2, 2)

    # Plot heatmap
    sns.heatmap(
    matrix,
    annot=annotations,
    fmt='',
    cmap="Blues",
    cbar=False,
    xticklabels=labels,
    yticklabels=labels,
    ax=ax,
    annot_kws={"fontsize": 14}  #  Increase this value to make text bigger
)


    ax.set_title(title, fontsize=16)
    ax.set_xlabel("Predicted", fontsize=16)
    ax.set_ylabel("True", fontsize=16)
    ax.tick_params(labelsize=16)

plt.suptitle("Confusion Matrices of Unsupervised Models", fontsize=18)
plt.tight_layout()
plt.savefig("conf_matrices_unsupervised.png", dpi=300)  # High-resolution save
plt.show()
```

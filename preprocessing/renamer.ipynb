```python
import os

# Define the folder path
folder_path = "/Users/berkeayyildizli/ENS_Graduation_Project/images"

# Iterate through all files in the folder
for filename in os.listdir(folder_path):
    old_path = os.path.join(folder_path, filename)
    new_path = os.path.join(folder_path, filename + ".csv")

    # Check if it's a file and does not already have .csv
    if os.path.isfile(old_path) and not filename.endswith(".csv"):
        os.rename(old_path, new_path)
        print(f"Renamed: {filename} -> {filename}.csv")

print("Renaming complete.")
```

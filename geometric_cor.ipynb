```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("thermal_geometric_features.csv")

# Drop the 'Frame' column (not a numeric feature)
df = df.drop(columns=["Frame"])

# Select only geometric features (adjust names if needed)
geometric_features = [
    "Num_Blobs", "Largest_Blob_Area", "Mean_Blob_Area",
    "Eccentricity", "Extent", "Solidity", "Orientation"
]

# Compute and plot correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(df[geometric_features].corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Geometric Features", fontsize=16)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

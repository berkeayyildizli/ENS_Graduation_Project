```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("labeled_thermal_features.csv")

# Drop the "Frame" column as it is not numerical
df.drop(columns=["Frame"], inplace=True)

# Display basic information
print("Dataset Info:")
print(df.info())

# Summary statistics
print("\nSummary Statistics:")
print(df.describe())

# Check for missing values
print("\nMissing Values:")
print(df.isna().sum())

# Check class distribution (Porosity Label)
plt.figure(figsize=(6, 4))
sns.countplot(x=df["Porosity Label"], palette="Set2")
plt.title("Class Distribution of Porosity Labels")
plt.xlabel("Porosity Label (0 = Normal, 1 = Porosity)")
plt.ylabel("Count")
plt.show()

import matplotlib.pyplot as plt
import seaborn as sns

# Correlation heatmap
plt.figure(figsize=(12, 8))
heatmap = sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)

# Balk ve etiketlerin yaz boyutlarn ayarla
plt.title("Feature Correlation Heatmap", fontsize=16)
plt.xticks(fontsize=14, rotation=45)
plt.yticks(fontsize=14, rotation=0)

plt.tight_layout()
plt.show()


# Histogram of temperature-related features
temp_features = ["Min_Temp", "Max_Temp", "Mean_Temp", "Std_Temp", "Median_Temp"]
df[temp_features].hist(figsize=(12, 8), bins=30)
plt.suptitle("Temperature Feature Distributions", fontsize=14)
plt.show()

# Boxplot of features grouped by Porosity Label
plt.figure(figsize=(12, 6))
sns.boxplot(x="Porosity Label", y="Mean_Temp", data=df, palette="muted")
plt.title("Mean Temperature vs Porosity Label")
plt.show()

sns.pairplot(df, hue="Porosity Label", diag_kind="kde", palette="husl")
plt.suptitle("Pairplot of Features", y=1.02)
plt.show()
```
